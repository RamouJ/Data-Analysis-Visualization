```python
import pandas as pd
import numpy as np

#function to decide if row is good data or bad
def isDataGood(numStudents,numVaccinated):
    isGood =  "This row has good data"
    if pd.isna(numStudents) or pd.isna(numVaccinated):
        isGood="numStudents or numVaccinated is not a valid number"
    if numStudents > 200: 
         isGood="numStudents is too big"
    if numStudents < 20: 
         isGood="numStudents is too small"
    if numStudents <  numVaccinated: 
         isGood="numStudents less than numVaccinated"
    if numVaccinated < 0: 
         isGood="numVaccinated can't be negative"
 
    return isGood
    

# read vaccination data
vaccines = pd.read_csv('vaccines.csv')
#print(vaccines.head(10))
print("****Original Rows, Cols****")
print(vaccines.shape)

#Rename columns
vaccines = vaccines.rename(columns={'Population':'Number of Students','Vaccinated number':'Students Vaccinated'})
print("****New Column Names****")
print(vaccines.columns)
print("****Add a Data Problem Column****")
vaccines['Data Problem'] = "not yet evaluated"
print(vaccines.columns)

#iterate over all the rows in the data set lookign for data problems
for i in vaccines.index:
        numStudents = vaccines.loc[i,'Number of Students']
        numVaccinated = vaccines.loc[i,'Students Vaccinated']
        vaccines.loc[i,'Data Problem'] = isDataGood(numStudents,numVaccinated)

badDataRows = vaccines[(vaccines["Data Problem"]!="This row has good data")]
print("****There are {} bad rows in the data set****".format(badDataRows.shape[0]))
print(badDataRows)

#get rid of the bad rows
vaccines = vaccines[(vaccines["Data Problem"]=="This row has good data")]
print("****There are {} rows after removing the bad ones****".format(vaccines.shape[0]))

#Fix data types
print("****Data types before****")
print(vaccines.dtypes)
vaccines = vaccines.astype({"Number of Students": int, "Students Vaccinated": int})
print("****Data types after****")
print(vaccines.dtypes)

#Remove extra column
vaccines = vaccines.filter(items=['School','Number of Students','Students Vaccinated'])
print("****There are {} rows and {} columns in the final data set****".format(vaccines.shape[0], vaccines.shape[1]))

 
#Save prepped file
newFileName = "vaccines_prepped.csv"
print("****Saved prepped DataFrame as {}****".format(newFileName))

vaccines.to_csv(newFileName, index=False)

```

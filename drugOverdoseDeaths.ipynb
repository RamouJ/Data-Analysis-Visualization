```python
import pandas as pd
import numpy as np

def translateState(state):
    result = {
        'CA': "California",
        'WA': "Washington"
    }
    return result.get(state,"state not found")

overdose = pd.read_csv('vsrr-provisional-drug-overdose-death-counts.csv')
print("****Rows in Original Table: {",overdose.shape[0],"}****")
overdose = (overdose[(overdose["State"]=="CA") | (overdose["State"]=="WA")])
overdose = overdose[overdose["Indicator"]=="Number of Drug Overdose Deaths"]
overdose = overdose[overdose["Predicted Value"]!="no data"]
print("****Rows in Filtered and Fixed: {",overdose.shape[0],"}****")
overdose['Predicted Value'] = overdose['Predicted Value'].astype(int)
overdose['Data Value'] = overdose['Data Value'].astype(int)
print(overdose.dtypes)
difference = (overdose['Predicted Value'] - overdose['Data Value'])
overdose['Difference'] = difference
overdose = overdose.filter(items=['State','Month','Year','Data Value','Predicted Value','Difference'])
overdose = overdose.rename(columns={'Data Value':'Observed','Predicted Value':'Predicted'})
overdose['State'] = overdose['State'].apply(translateState)
overdose = overdose.sort_values(by=["State","Predicted"], ascending=[False,False])
print(overdose)
```

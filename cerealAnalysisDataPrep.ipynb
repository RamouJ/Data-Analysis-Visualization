```python
import pandas as pd
import numpy as np

cereals = pd.read_csv("cereals2.csv")
print("****Original Rows, Cols****")
print(cereals.shape)
print("****New Column Names")
cereals = cereals.filter(items=['name','mfr','protein','carbo','sugars','calories'])
cereals = cereals.rename(columns={'name':'Cereal','mfr':'Company','protein':'Protein','carbo':'Carbs','sugars':'Sugar','calories':"Calories"})
print(cereals.columns)
numRows = cereals.shape[0]
cereals = cereals.iloc[1:numRows]
print("****Rows, Cols after Removing Row 1****")
print(cereals.shape)
print("****NA Data in Table****")
print(cereals.isna())
print("****Rows, Cols after Removing Non Numbers****")
cereals = cereals.dropna(subset=['Cereal','Company','Protein','Carbs','Sugar','Calories'])
print(cereals.shape)
print("****Data Types Before Conversion")
print(cereals.dtypes)
cereals['Protein'] = cereals['Protein'].astype(int)
cereals['Carbs'] = cereals['Carbs'].astype(float)
cereals['Sugar'] = cereals['Sugar'].astype(int)
print("****Data Types After Conversion")
print(cereals.dtypes)
print("****Rows, Cols after Numbers Out of Range****")
cereals = cereals[(cereals["Sugar"]>=0) & (cereals["Carbs"]>=0)]
print(cereals.shape)
cereals.to_csv("cereals2_prepped.csv", index=False)
```
